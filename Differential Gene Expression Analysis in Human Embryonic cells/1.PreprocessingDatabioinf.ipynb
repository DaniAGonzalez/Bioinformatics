```python
# Loading data as df
import pandas as pd
df = pd.read_csv('/content/41594_2013_BFnsmb2660_MOESM31_ESM.csv')
```

```python
df.head(5)
```

```python
# Filtering columns
columns_to_keep = ["Gene_ID", "Transcript_ID"]
columns_to_keep += [col for col in df.columns if "Average" in col]
df_filtered = df[columns_to_keep]
```

```python
#Filtering to retain only these columns
filtered_df = df[columns_to_keep]
```

```python
filtered_df.head(5)
```

```python
shape= filtered_df.shape
print(shape)
```

```python
#Checking value types
data_types = filtered_df.dtypes
print(data_types)
```

```python
# Checking null values and filling with mean imputation
nan_counts = filtered_df.isnull().sum()
print(f"Missing values in each column:", nan_counts)


```

```python
# Checking duplicated
duplicated_rows = filtered_df.duplicated().sum()
print(f"Duplicated values:, {duplicated_rows}")
```

```python
num_cols = ['Average Oocyte(RPKM)', 'Average Zygote(RPKM)',
            'Average 2-cell embryo(RPKM)', 'Average 4-cell embryo(RPKM)',
            'Average 8-cell embryo(RPKM)', 'Average Morulae(RPKM)',
            'Average Trophectoderm(RPKM)', 'Average Primitive endoderm(RPKM)',
            'Average epiblast(RPKM)','Average hESC passage#0(RPKM)',
            'Average hESC passage#10(RPKM)']

# Outliers detection
Q1 = filtered_df[num_cols].quantile(0.25)
Q3 = filtered_df[num_cols].quantile(0.75)
IQR = Q3-Q1

#Checking values not in IQR
outliers = ((filtered_df[num_cols] < (Q1 - 1.5 * IQR)) | (filtered_df[num_cols] > (Q3 + 1.5 * IQR)))
print(outliers.sum())
```

```python
# Saving the filtered_df
filtered_df.to_csv('filtered_df.csv', index=False)
```
